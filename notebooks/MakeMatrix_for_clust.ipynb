# Making matrix for clustering

```python
''' 
You need to have *fam70_test.txt* ~ lable-file, generated by Creating_sample.ipynb
"0" - different families, "1" - the same families; for fam70_test
'''
import numpy as np
import pandas as pd

st_list = []
with open("fam70_test.txt", 'r') as fname: 
    for string in fname:
        st_list.append(string.strip())
zero_matrix = np.zeros(len(st_list)**2).reshape(len(st_list), len(st_list))
```

```python
# Checking if sequences belong to the same family
for str_i in range(len(st_list)):
    str_item_fam = int(st_list[str_i].split(":")[1])
    str_item_id = st_list[str_i].split(":")[0]
    
    for col_i in range(len(st_list)):
        col_item_fam = int(st_list[col_i].split(":")[1])
        col_item_id = st_list[col_i].split(":")[0]
        
        if str_item_id == col_item_id:
            zero_matrix[col_i][str_i] += float(1)

        elif str_item_fam == col_item_fam:
            zero_matrix[col_i][str_i] += float(1)
```

```python
# Save your vector into file to use it later
np.save('./v7_test_62_matrix.npy', zero_matrix)
```
